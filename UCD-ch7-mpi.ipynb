- mpi4py jobs must be launched with _mpirun_ or _mpiexec_
- each rank launches its own independent python interpreter each interpreter only has access to files and libraries available locally to it, unless distributed to the ranks communication is handled by your MPI layer
- any function outside of an if block specifying a rank is assumed to be global

Standard MPI commands:

- Get_rank()
- Get_size()

- send(data, dest, tag) 
- recv(data, source, tag)
- request = isend()
- irecv
- bcast(data, root)
- scatter(data, root)
- gather(data, root)
- allgather(data)
- reduce(data, operation, root)
- allreduce(data, operation)
- Barrier()
- Disconnect() or Finallize()

Most of the commands come in two ways:
    
- **lower-case** methods are for Python generic objects
- **upper case** for more general buffers that must be specified by type in a list/tuple  [data, MPI.double] or [data, count, MPI.DOUBLE] (this way it's also faster)

### Point to point communication

```python
import numpy as np
from mpi4py import MPI
comm = MPI.COMM_WORLD
rank, size = comm.Get_rank(), comm.Get_size()

# pass Python generic objects
if rank == 0:
    data = {'a': 7, 'b': 3.14}
    print 'rank %i sends data:'%rank, data
    comm.send(data, dest=1, tag=11)
elif rank == 1:
    data = comm.recv(source=0, tag=11)
    print 'rank %i got data:'%rank, data

#    
# buffer-general objects
#
import numpy
# pass explicit MPI datatypes
if rank == 0:
    data = numpy.arange(10, dtype='i')
    print 'rank %i sends data:'%rank, data
    comm.Send([data, MPI.INT], dest=1, tag=77)
elif rank == 1:
    data = numpy.empty(10, dtype='i')
    comm.Recv([data, MPI.INT], source=0, tag=77)
    print 'rank %i got data:'%rank, data

# automatic MPI datatype discovery
if rank == 0:
    data = numpy.arange(10, dtype=numpy.float64)
    print 'rank %i sends data:'%rank, data
    comm.Send(data, dest=1, tag=13)
elif rank == 1:
    data = numpy.empty(10, dtype=numpy.float64)
    comm.Recv(data, source=0, tag=13)
    print 'rank %i got data:'%rank, data
```

### Non-blocking (asynchronous) send & recieve

- Standard send and recieve don't return until the send or the receive has been completed
- This can easily lead to a deadlock if used brainless.

Innocent but very wrong example:
    
    ... block of code common for all processes ...
    comm.send(data, dest, tag)
    data = comm.recv(source, tag)
    
- Isend and Irecv
- barrier needed before the next common task to prevent one process is ahead in the code of the other

```python
# non-blocking send & recieve

import numpy as np
from mpi4py import MPI
comm = MPI.COMM_WORLD
rank, size = comm.Get_rank(), comm.Get_size()

assert comm.size == 2

if comm.rank == 0:
    array1 = np.arange(10, dtype = 'f')
    array2 = np.empty(10, dtype = 'f')
    target = 1
else:
    array1 = np.ones(10, dtype = 'f')
    array2 = np.empty(10, dtype = 'f')
    target = 0

print 'rank %i sends data:'%rank, array1
request = comm.Isend([array1, MPI.FLOAT], dest = target)
comm.Recv([array2, MPI.FLOAT], source = target)
request.Wait()
# comm.Request.Waitall([request1, request2])
print 'rank %i got data:'%rank, array2
```

###Pi by Monte Carlo method

$ \pi = 4\frac{N (x^2 + y^2 < 1)}{N}$ for $x \in (0,1), y \in (0,1)$

```python
import numpy as np

def get_pi_approx(N=1e6):
    """ Calculate approximation of pi by random throws
        in the unit square measuring ratio of hits in the unit circle
    """
    x, y = np.random.random((2, N))
    r2 = x**2 + y**2
    Ns = np.sum(r2 <= 1.0)
    pi_approx = Ns / N * 4
    return pi_approx

print get_pi_approx()
```

```python
import numpy as np
from mpi4py import MPI
comm = MPI.COMM_WORLD
rank, size = comm.Get_rank(), comm.Get_size()

pi = get_pi_approx()

if rank != 0:
    comm.Send(pi, dest=0)
else:
    pi_vec = np.empty((size,1), dtype=np.float)
    pi_vec[0] = pi
    for source in xrange(1, size):
        comm.Recv(pi_vec[source], source=source)
    # Calculate arithmetic average of results
    pi_final = np.sum(pi_vec) / size
    print 'pi approximation: ', pi_final
    print 'pi reference:     ', np.pi
```

### Mandelbrot

```python
# Mandelbrot Set

def mandelbrot(x, y, maxit):
    c = x + y*1.j
    z = 0. + 0.j
    it = 0
    while abs(z) < 2 and it < maxit:
        z = z**2 + c
        it += 1
    return it
```

```python
# ￼Mandelbrot Set - sequential

x1, x2 = -2.0, 1.0
y1, y2 = -1.0, 1.0
w, h   = 1500, 1000
maxit= 127

import numpy
import matplotlib.pyplot as plt

C = numpy.zeros([h, w], dtype='i')
dx=(x2-x1)/w
dy=(y2-y1)/h
for i in range(h):
    y = y1 + i * dy
    for j in range(w):
        x = x1 + j * dx
        C[i, j] = mandelbrot(x, y, maxit)

plt.imshow(C, aspect='equal')
plt.spectral()
plt.show()
```

```python
# ￼Mandelbrot Set - parallel, block

from mpi4py import MPI
import numpy

comm = MPI.COMM_WORLD
size = comm.Get_size()
rank = comm.Get_rank()

# number of rows to compute here
N=h//size+(h%size>rank)
# first row to compute here
start = comm.scan(N)-N # omitting the operation means SUM
# array to store local result
Cl = numpy.zeros([N, w], dtype='i')
# compute owned rows
dx = (x2 - x1) / w
dy = (y2 - y1) / h
for i in range(N):
    y = y1 + (i + start) * dy
    for j in range(w):
        x = x1 + j * dx
        Cl[i, j] = mandelbrot(x, y, maxit)
# gather results at root (process 0)
counts = comm.gather(N, root=0)
C=None
if rank == 0:
    C = numpy.zeros([h, w], dtype='i')
# special way how to create a contigous array of INTs just to send them in a message
rowtype = MPI.INT.Create_contiguous(w)
rowtype.Commit()
comm.Gatherv(sendbuf=[Cl, MPI.INT], recvbuf=[C, (counts, None), rowtype], root=0)
rowtype.Free()

if comm.rank == 0:
    import matplotlib.pyplot as plt
    plt.imshow(C, aspect='equal')
    plt.spectral()
    plt.show()
```

###Calculate Pi in parallel using atan
$$\pi = 4 \int^1_0 \frac{1}{1+x^2}dx \approx \frac{1}{n} 4 \sum_0^{n-1}\frac{1}{1+(\frac{i + 0.5}{n})^2}$$

```python
import math

def compute_pi(n):
    h = 1.0 / n
    s = 0.0
    for i in range(n):
        x = h * (i + 0.5)
        s += 4.0 / (1.0 + x**2)
    return s * h

n=10
pi = compute_pi(n)
error = abs(pi - math.pi)
print ("pi is approximately %.16f, error is %.16f" % (pi, error))
```

```python
# vector x matrix multiplication
from mpi4py import MPI
import numpy

def matvec(comm, A, x):
    m = A.shape[0] # local rows
    p = comm.Get_size()
    xg = numpy.zeros(m*p, dtype='d')
    comm.Allgather([x,  MPI.DOUBLE],
                   [xg, MPI.DOUBLE])
    y = numpy.dot(A, xg)
    return y
```

```python
from mpi4py import MPI
comm = MPI.COMM_WORLD
import random

rank, size = comm.rank, comm.size

def compute_pi(nsamples):
    inside = 0.0
    for i in xrange(nsamples):
        x = random.random()
        y = random.random()
        if (( x*x) + (y*y) < 1.0):
            inside += 1

    mypi = (inside) /float(nsamples)
    pi = (4.0/size) * comm.allreduce(mypi, op=MPI.SUM)
    return pi

pi = compute_pi(2000000)/size
if rank == 0:
    print "Computed values of of Pi (on %i cores) = %.6f"%(size, pi) 
```

```python
# ping pong

from mpi4py import MPI
comm = MPI.COMM_WORLD
sendmsg = [comm.rank]*3
right = (comm.rank + 1 ) % comm.size
left = (comm.rank - 1 ) % comm.size
    
req1 = comm.isend(sendmsg, dest= right)
req2 = comm.isend(sendmsg, dest = left)
lmsg = comm.recv(source = left)
rmsg = comm.recv(source = right)
    
MPI.Request.Waitall([req1, req2])
assert lmsg == [left]*3
assert rmsg == [right]*3
```

```python
# matrix multiplication
from mpi4py import MPI
import numpy as np
from time import time

comm = MPI.COMM_WORLD
size, rank = comm.size, comm.rank

# get a dimension which is multiple of # of cpus
x = int(np.ceil(4000.0/size)*size)
y = 2000
step = x/size

# scatter destination
domain = np.zeros((step,y),dtype='d')
# gather destination
c = np.zeros((x,x),dtype='d')
if rank==0:
    # two matrices to multiply
    a = np.random.uniform(size=(x,y)).astype('d')
    b = np.random.uniform(size=(y,x)).astype('d')
else:
    # target buffers
    a = np.zeros((x,y),dtype='d')
    b = np.zeros((y,x),dtype='d')

# broadcast 2nd matrix
comm.Bcast([b,MPI.DOUBLE])

comm.Scatter([a,MPI.DOUBLE],[domain,MPI.DOUBLE])
ans = np.dot(domain,b)
comm.Gather([ans,MPI.DOUBLE],[c,MPI.DOUBLE])

if rank==0:
    single_c = np.dot(a,b)
    print np.alltrue(abs(c-single_c)<1e-5)
```
